## 1. Manipulation des données avec Pandas/ NumPy

### a) Lire le CSV et explorer les données (shape, head, info)

```python
# import des bibliothèques
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('movies.csv')
```

```python
# CONVERSION DES COLONNES EN TYPE NUMÉRIQUE

print("=== AVANT CONVERSION ===")
print(df[['budget', 'revenue', 'vote_average', 'vote_count']].dtypes)
print()

# Colonnes à convertir en numérique
colonnes_a_convertir = ['budget', 'revenue', 'vote_average', 'vote_count', 'runtime', 'popularity']

for col in colonnes_a_convertir:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
        print(f"✅ {col} converti en numérique")

print()
print("=== APRÈS CONVERSION ===")
print(df[['budget', 'revenue', 'vote_average', 'vote_count']].dtypes)
print()
print("✅ Toutes les colonnes numériques sont converties !")
```

```python
# Afficher le nombre de lignes et de colonnes
print("Taille du dataset :")
print(df.shape)
print()

# Plus détaillé
print(f"Nombre de films : {df.shape[0]}")
print(f"Nombre de colonnes : {df.shape[1]}")
```

```python
# Afficher les 5 premiers films
print("Les 5 premiers films :")
df.head()
```

```python
# Afficher les infos sur chaque colonne 
df.info()
```

### b) Nettoyer les valeurs manquantes (budget, revenue, rating)

```python
# Afficher l'état AVANT nettoyage
print("Valeurs manquantes :")
print(df[['budget', 'revenue', 'vote_average']].isnull().sum())
```

```python
# Budhet n'a pas de valeur manquante 

# Calculer la moyenne du revenue
moyenne_revenue = df['revenue'].mean()

# Remplacer les valeurs manquantes par la moyenne
df['revenue'] = df['revenue'].fillna(moyenne_revenue)

print(f"✅ Revenue nettoyé avec la moyenne du revenue qui est de {round(moyenne_revenue, 2)}")
```

```python
# Compter combien de films avant
avant = df.shape[0]

# Supprimer les films sans vote_average
df = df.dropna(subset=['vote_average'])

# Compter combien après
apres = df.shape[0]

print(f"✅ Vote_average nettoyé")
print(f"Films supprimés : {avant - apres}")
print(f"Films restants : {apres}")
```

```python
# Remplacer les 0 de budget par la moyenne
moyenne_budget = df[df['budget'] > 0]['budget'].mean()
df.loc[df['budget'] == 0, 'budget'] = moyenne_budget

print(f"✅ Budgets à 0 remplacés par la moyenne : {moyenne_budget:,.2f}")
```

```python
# Calculer la moyenne des revenues > 0
moyenne_revenue = df[df['revenue'] > 0]['revenue'].mean()

print(f"Moyenne des revenues (hors 0) : {moyenne_revenue:,.2f}")

# Remplacer les 0 par la moyenne
nombre_zeros = (df['revenue'] == 0).sum()
df.loc[df['revenue'] == 0, 'revenue'] = moyenne_revenue

print(f"✅ {nombre_zeros} valeurs de 0 remplacées par la moyenne")
```

### c) Créer de nouvelles colonnes : profit = revenue - budget.

```python
# Convertir revenue en nombre (version sans warning)
df.loc[:, 'revenue'] = pd.to_numeric(df['revenue'], errors='coerce')

print("✅ Budget et revenue convertis en nombres")
```

```python
# Créer la colonne profit (version sans warning)
df.loc[:, 'profit'] = df['revenue'] - df['budget']

print("✅ Colonne profit créée")
```

```python
# Afficher les 5 premières lignes
df.head()
```

### d) Filtrer / trier / grouper : top 10 films par profit, rating, etc.

```python
# Trier par profit décroissant et prendre les 10 premiers
top_10_profit = df.nlargest(10, 'profit')

# Afficher seulement les colonnes importantes
print("Top 10 films par profit :")
top_10_profit[['title', 'profit', 'budget', 'revenue']]
```

```python
# Trier par vote_average décroissant et prendre les 10 premiers
top_10_rating = df.nlargest(10, 'vote_average')

# Afficher seulement les colonnes importantes
print("Top 10 films par rating :")
top_10_rating[['title', 'vote_average', 'vote_count']]
```

```python
# Convertir profit en numérique
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')

# Trier par budget décroissant et prendre les 10 premiers
top_10_budget = df.nlargest(10, 'budget')

# Afficher les colonnes importantes
print("Top 10 films par budget :")
top_10_budget[['title', 'budget', 'revenue', 'profit']]
```

```python
# Sauvegarder le DataFrame nettoyé dans un nouveau fichier CSV
df.to_csv('movies_clean.csv', index=False)

print("✅ Fichier movies_clean.csv sauvegardé !")
```
